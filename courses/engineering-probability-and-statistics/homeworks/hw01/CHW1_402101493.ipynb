### 1.1 Random Walk

Assume a fair coin has `1` on one side and `-1` on the other. Therefore the random variable $X$ has a Bernoulli distribution defined as:
$$
X \sim 2 \times \mathsf{Ber}(\frac{1}{2}) - 1
$$

But you will be tossing the coin multiple times, which means the random variable is defined as $X_t$; $X_1$ is the result of the first toss, $X_2$ is the result of the second toss, etc.

Now, assume a person is standing on a one-dimensional line, at the origin ($L=0$). At each discrete time $t$, he will toss a coin with the outcome $X_t$, and will move forward 1 unit if the result is `1`, and will move backwards 1 unit if the result is `-1`. This is a classic example of a "Random Walk". We are interested in simulating the trajectory of this specific walk.

The formula for finding the location $L_t$ of the person is:
$$
L_t = \sum^t_{i=1} X_i
$$

#### 1.1.1 Visualization

For $t$ from $1$ to $T = 2000$, simulate the trajectory of this person's walk, $N=3$ times. Your output should be a plot with 3 subplots. In each subplot, the x-axis should be $t$ and the y-axis must be the location $L$ at that time.

```python
import numpy as np
import matplotlib.pyplot as plt

T, N = 2000, 3

walks = [np.cumsum(np.random.choice([-1, 1], size=T)) for _ in range(N)]

for idx, L in enumerate(walks, start=1):
    plt.plot(np.arange(1, T + 1), L, label=f'Walk {idx}')

plt.xlabel('Time (t)')
plt.ylabel('Location (L)')
plt.legend()
plt.show()
```

#### 1.2.1 Normal Distribution

In this question we want to visualize the distrubtions of different normal distributions.

The PDF of a normal distribution $N(\mu, \sigma^2)$ with parameters $\mu$ and $\sigma$ is defined as:
$$
p_X(x) = \frac{1}{\sigma \sqrt{2 \pi}} e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^2}
$$

Another notation for $p_X(x)$ is $f_X(x)$, so do not be confused, they are the same thing!

Plot the distributions of the three random variables defined below in 3 subplots (you simply have to draw the functions):

\begin{align}
Z_1&\sim N(0,1) \\
Z_2&\sim N(5,1) \\
Z_3&\sim N(0,4)
\end{align}

Include the values of $\mu$ and $\sigma^2$ in the title of each subplot.

```python
import numpy as np
import matplotlib.pyplot as plt

params = [(0, 1), (5, 1), (0, 2)] 
x = np.linspace(-10, 10, 1000)

for idx, (mu, sigma) in enumerate(params, start=1):
    p = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    plt.plot(x, p, label=f'Distribution {idx}')

plt.xlabel('x')
plt.ylabel('PDF')
plt.legend()
plt.show()
```

What does changing the parameter $\mu$ and $\sigma^2$ do?


A change in the mean causes a shift in the graph, while a change in variance affects its compression or spread. If the variance increases, the graph becomes wider and more dispersed, whereas if the variance decreases, the graph becomes more compressed.

### 1.4 2D Random Processes

#### 1.4.1 The Rayleigh Distribution

The Rayleigh Distribution is defined as:

$$
\begin{cases}
p_X(x) = \frac{x}{\sigma^2}e^{-\frac{x^2}{2\sigma^2}} & \text{for}\ x > 0 \\
0 & o.w
\end{cases}
$$

Plot this distribution for the $\sigma^2$ values given below, on the same plot with a legend. [More information here](https://en.wikipedia.org/wiki/Rayleigh_distribution)

$$
\begin{cases}
\sigma^2_1 = 1\\
\sigma^2_2 = 4\\
\sigma^2_3 = 9
\end{cases}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

sigma_values = [1, 4, 9]
x = np.linspace(0, 10, 1000)

p_values = [
    (x / sigma) * np.exp(-((x ** 2) / (2 * sigma))) 
    for sigma in sigma_values
]

for p in p_values:
    plt.plot(x, p)

plt.xlabel('x')
plt.ylabel('PDF')
plt.legend([f'Sigma = {sigma}' for sigma in sigma_values])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def Game1(n):
    return np.sum([6 in np.random.randint(1, 7, 4) for _ in range(n)]) / n

def Game2(n):
    return np.sum([
        any(a == 6 for a in np.random.randint(1, 7, 24)) and
        any(b == 6 for b in np.random.randint(1, 7, 24))
        for _ in range(n)
    ]) / n

def simulate_games(n_values):
    return [(n, Game1(n), Game2(n)) for n in n_values]

n_values = [10, 100, 1000]

results = simulate_games(n_values)

w1 = [result[1] for result in results]
w2 = [result[2] for result in results]

plt.plot(n_values, w1, label="Game 1")
plt.plot(n_values, w2, label="Game 2")
plt.xlabel("n")
plt.ylabel("W")
plt.legend()
plt.show()

```

### 3.1 Circles are the key!
First let's solve for one simpler version of this question:

> Three points are chosen at random on circumference of a circle. What is the probability that the center of the circle lies inside the triangle whose vertices are at the three points?'

First write a function to generate a random point. You can achieve this by introducing an angle an returning the $sin$ and $cos$ values of that angle.

```python
import numpy as np
from typing import Tuple

def random_point_on_circle() -> Tuple[float, float]:
    theta = np.random.uniform(0, 2 * np.pi) 
    x = np.cos(theta)  
    y = np.sin(theta)  
    return x, y
```

Now write a function to determine if the center of the circle lies inside the triangle formed by the three points. Keep an eye on the comments!

```python
import numpy as np
from typing import List, Tuple

def is_center_inside_triangle(points: List[Tuple[float, float]]) -> bool:

    
    def area(a: Tuple[float, float], b: Tuple[float, float], c: Tuple[float, float]) -> float:
        return 0.5 * abs(a[0] * (b[1] - c[1]) + b[0] * (c[1] - a[1]) + c[0] * (a[1] - b[1]))
    
    main_area = area(points[0], points[1], points[2])
    
    area1 = area(points[0], points[1], (0, 0))
    area2 = area(points[0], (0, 0), points[2])
    area3 = area((0, 0), points[1], points[2])
    

    return np.isclose(main_area, area1 + area2 + area3)
```

Now run the simulation!

```python
import numpy as np
from typing import Tuple, List

class CircleSimulation:
    def __init__(self, num_trials: int):
        self.num_trials = num_trials

    def random_point_on_circle(self) -> Tuple[float, float]:

        teta = np.random.uniform(0, 2 * np.pi) 
        x = np.cos(teta) 
        y = np.sin(teta) 
        return x, y

    def area(self, a: Tuple[float, float], b: Tuple[float, float], c: Tuple[float, float]) -> float:
   
        return 0.5 * abs(a[0] * (b[1] - c[1]) + b[0] * (c[1] - a[1]) + c[0] * (a[1] - b[1]))

    def is_center_inside_triangle(self, points: List[Tuple[float, float]]) -> bool:
   
        A = self.area(points[0], points[1], points[2]) 
        A1 = self.area(points[0], points[1], (0, 0)) 
        A2 = self.area(points[0], (0, 0), points[2])
        A3 = self.area((0, 0), points[1], points[2])

        return np.isclose(A, A1 + A2 + A3) 

    def simulate(self) -> float:

        count = 0
        for _ in range(self.num_trials):
            points = [self.random_point_on_circle() for _ in range(3)] 
            if self.is_center_inside_triangle(points):
                count += 1
        return count / self.num_trials

num_trials = 10000
simulation = CircleSimulation(num_trials)
probability_2d = simulation.simulate()

print(f"Probability in 2D: {probability_2d}")
```

The approach for this one is the same. We have to write a function to check if the tetrahedron contains the center.

```python
import numpy as np
from typing import List, Tuple

def calculate_signed_volume(points: List[Tuple[float, float, float]]) -> float:

    matrix = np.hstack([points, np.ones((4, 1))])  
    return np.linalg.det(matrix) / 6 

def is_center_inside_tetrahedron(points: List[Tuple[float, float, float]]) -> bool:
  
    center = np.array([0.0, 0.0, 0.0])
    
    volume1 = calculate_signed_volume([center, points[1], points[2], points[3]])
    volume2 = calculate_signed_volume([center, points[0], points[2], points[3]])
    volume3 = calculate_signed_volume([center, points[0], points[1], points[3]])
    volume4 = calculate_signed_volume([center, points[0], points[1], points[2]])
    
    return np.sign(volume1) == np.sign(volume2) == np.sign(volume3) == np.sign(volume4)
```

Run the simulation:

```python
import numpy as np
from typing import Tuple, List

def generate_random_coordinates() -> Tuple[float, float, float]:
    u, v = np.random.random(2)
    theta = 2 * np.pi * u
    phi = np.arccos(2 * v - 1)
    x = np.sin(phi) * np.cos(theta)
    y = np.sin(phi) * np.sin(theta)
    z = np.cos(phi)
    return x, y, z

def calculate_signed_volume_of_tetrahedron(vertices: List[Tuple[float, float, float]]) -> float:
    extended_matrix = np.hstack([vertices, np.ones((4, 1))])
    return np.linalg.det(extended_matrix) / 6

def is_origin_inside_tetrahedron(vertices: List[Tuple[float, float, float]]) -> bool:
    origin = np.array([0.0, 0.0, 0.0])
    tetrahedron_volumes = [
        calculate_signed_volume_of_tetrahedron([origin, vertices[1], vertices[2], vertices[3]]),
        calculate_signed_volume_of_tetrahedron([origin, vertices[0], vertices[2], vertices[3]]),
        calculate_signed_volume_of_tetrahedron([origin, vertices[0], vertices[1], vertices[3]]),
        calculate_signed_volume_of_tetrahedron([origin, vertices[0], vertices[1], vertices[2]])
    ]
    return all(volume > 0 for volume in tetrahedron_volumes) or all(volume < 0 for volume in tetrahedron_volumes)

def run_3d_probability_simulation(num_trials: int) -> float:
    inside_count = 0
    for _ in range(num_trials):
        vertices = [generate_random_coordinates() for _ in range(4)]
        if is_origin_inside_tetrahedron(vertices):
            inside_count += 1
    return inside_count / num_trials


trials = 10000
probability_3d = run_3d_probability_simulation(trials)
print(f"Probability in 3D: {probability_3d}")
```
